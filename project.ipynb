```python
datafig_1.rename(columns={'21stCenturyTheilSenSlope_m_per_year': 'Early21stCenturyTheilSenSlope_m_per_year'}, inplace=True)
aquifers_merged = datafig_1.merge(datafig_3, on='Early21stCenturyTheilSenSlope_m_per_year', how='outer')
```

```python
aquifers_merged
```

```python
#Rango de coordenadas: 42.56 3.57 (Noreste)
#                      35.12 0.95 (Sureste)
#                      36.05 -11.6 (Suroeste)
#                      44.27 -10.52 (Noroeste)
```

```python
aquifers_spain = aquifers_merged.query('44.27 > Latitude > 35.12')
aquifers_spain = aquifers_spain.query('3.57 > Longitude > -11.6')


# aquifers_spain.drop_duplicates(subset=['StnID'], inplace=True)
aquifers_spain
```

```python
sf_map = folium.Map()

aquifers = folium.map.FeatureGroup()


for lat, lng, label in zip(aquifers_spain["Latitude"], aquifers_spain["Longitude"], aquifers_spain['AquiferSystem']): 
    
    aquifers.add_child(folium.Marker(location=[lat, lng], popup=label))


# Agrega incidentes al mapa
sf_map.add_child(aquifers)

sf_map
```
